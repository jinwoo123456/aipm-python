```python
plt.title("선 모양 지정")

plt.plot([1,2,3],[4,4,4],
         linestyle='solid',linewidth=5,
         solid_capstyle='butt',color='r',
         label='solid+butt')
plt.plot([1,2,3],[3,3,3],
         linestyle='solid',linewidth=10,
         solid_capstyle='round',color='b',
         label='solid+round')

plt.plot([1,2,3],[2,2,2],
         linestyle='dashed',linewidth=5,
         solid_capstyle='butt',color='g',
         label='dashed+butt')
plt.plot([1,2,3],[1,1,1],
         linestyle='solid',linewidth=10,
         solid_capstyle='round',color='y',
         label='dashed+round')
# dash_capstyle을 사용해서 선의 끝 모양 지정
# {'butt','projecting','round'}

plt.xlabel('X-Axis')
plt.xlabel('X-Yxis')
plt.axis([0.8,3.2,0.5,5.0])
plt.legend(loc='best',ncol=2, fontsize=14,shadow=True
) 
plt.tight_layout()
plt.show()
```

```python
# 수평선/수직선 표시

x = np.arange(0, 4, 0.5)

plt.plot(x, x + 1, 'bo')
plt.plot(x, x**2 - 4, 'g--')
plt.plot(x, -2*x + 3, 'r:')

plt.axhline(4.0, 0.1, 0.9, color='lightgray', linestyle='--', linewidth=4)
# axhline(y값, xmin, xmax) /xmin, xmax는 0에서 1사이 값. 0은 왼쪽 끝, 1은 오른쪽 끝
plt.hlines(-0.62, 1.0, 2.5, color='gray', linestyle='solid', linewidth=3)
# hlines(y, xmin, xmax) / 점 (xmin, y)에서 점 (xmax, y)를 따라 수평선을 표시

plt.show()
```

```python
# 수직선 그리기

x = np.arange(0, 4, 0.5)

plt.plot(x, x + 1, 'bo')
plt.plot(x, x**2 - 4, 'g--')
plt.plot(x, -2*x + 3, 'r:')

plt.axvline(4.0, 0.1, 0.9, color='lightgray', linestyle='--', linewidth=4)
# axvline(x값, ymin, ymax) 
plt.vlines(-0.62, 1.0, 2.5, color='gray', linestyle='solid', linewidth=3)
# ylines(x, ymin, ymax) / 점 (x, ymin)에서 점 (x, ymax)를 따라 수평선을 표시

plt.show()
```

```python
x = np.arange(3)
years = ['2018','2019','2020']
values = [100,400,900]

plt.bar(x, values)
plt.xticks(x, years) # x축과 눈금 레이블에 '2018','2019','2020'이 순서대로 표시

plt.show()
```

```python
np.random.seed(0)

n= 50
x = np.random.rand(n)
y = np.random.rand(n)
area = (10*np.random.rand(n))**2
colors = np.random.rand(n)

plt.scatter(x,y, s=area,c=colors, alpha=1,cmpa='Spectical')
plt.yticks(y, years) #  Y축의 눈금 레이블에 ‘2018’, ‘2019’, ‘2020’이 순서대로 표시

plt.show()
```

```python
data = np.random.randn(1000)
plt.hist(data, bins=30, histtype='stepfilled', edgecolor='black') 
                      # 선으로 히스토그램을 그리기
plt.title("Step Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(data, bins=30, orientation='horizontal', edgecolor='black') 
                  # 히스토그램을 수평 방향으로 그리기
plt.title("Horizontal Histogram")
plt.xlabel("Frequency")
plt.ylabel("Value")
plt.show()
```
