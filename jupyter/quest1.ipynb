```python
타이타닉호 승객을 "미성년자" "중년" "중년" "장년" "노인" 나이 그룹으로 바꾸기
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
titanic = sns.load_dataset("titanic")

titanic["category1"] = titanic["age"].apply(lambda x:'age'if x>20 else "child")

bins = [0,20,30,50,70]
labels = ["미성년자", "중년" , "장년","노인" ]
titanic["category2"] = pd.cut(titanic["age"],bins,labels=labels)
titanic
```
